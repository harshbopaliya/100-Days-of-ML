```python
import numpy as np
```

```python
import pandas as pd
```

```python
df = pd.read_csv('titanic.csv')
```

```python
df.head()
```

```python
df['number'].unique()
```

```python
fig = df['number'].value_counts().plot.bar()
fig.set_title('Passengers travelling with')
```

```python
df['number_numercial']= pd.to_numeric(df['number'],errors='coerce',downcast='integer')
```

```python
df['number_categorical']=np.where(df['number_numercial'].isnull(),df['number'],np.nan)
df.head()
```

```python
df['Ticket'].unique()
```

```python
df['cabin_num'] = df['Cabin'].str.extract('(\d+)')
df['cabin_cat'] = df['Cabin'].str[0]
df.head()
```

```python
fig = df['cabin_cat'].value_counts().plot.bar()
fig.set_title('cabin_cat travelling with')
```

```python
df['ticket_num'] = df['Ticket'].apply(lambda s: s.split()[-1])
df['ticket_num'] = pd.to_numeric(df['ticket_num'],
                                   errors='coerce',
                                   downcast='integer')

# extract the first part of ticket as category
df['ticket_cat'] = df['Ticket'].apply(lambda s: s.split()[0])
df['ticket_cat'] = np.where(df['ticket_cat'].str.isdigit(), np.nan,
                              df['ticket_cat'])

df.head(20)
```
