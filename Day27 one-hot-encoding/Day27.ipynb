# 3. OneHotCoding using sklearn

```python
from sklearn.model_selection import train_test_split
X_train,X_test,y_train,y_test = train_test_split(df.iloc[:,0:4],df.iloc[:,-1],test_size=0.2,random_state=2)
```

```python
X_train.head()
```

```python
from sklearn.preprocessing import OneHotEncoder
```

```python
ohe = OneHotEncoder(drop='first',sparse_output=False,dtype=np.int32)
```

```python
X_train_new =ohe.fit_transform(X_train[['fuel','owner']])
```

```python
X_test_new =ohe.transform(X_test[['fuel','owner']])
```

```python
X_train_new.shape
```

```python
X_combined = np.hstack((X_train[['brand','km_driven']].values,X_train_new))
```

```python
original_columns = ['brand', 'km_driven']
encoded_columns = ohe.get_feature_names_out(['fuel', 'owner'])
column_names = np.hstack((original_columns, encoded_columns))
```

```python
X_combined_df = pd.DataFrame(X_combined, columns=column_names)
```

# 4.OneHotEncoding with Top categories

```python
counts = df['brand'].value_counts()
```

```python
threshold=100
```

```python
repl = counts[counts <= threshold].index
```

```python
pd.get_dummies(df['brand'].replace(repl,'uncommon'))
```

```python
# Step 1: Count the frequency of each brand
brand_counts = df['brand'].value_counts()

# Step 2: Identify the top 5 brands
top_10_brands = brand_counts.index[:10]

# Step 3: Replace all other brands with 'Other'
df['brand_merged'] = df['brand'].apply(lambda x: x if x in top_10_brands else 'Other')

# Step 4: Plot the result
sns.histplot(df['brand_merged'], discrete=True)

# Add labels and title
plt.xlabel('Brand')
plt.ylabel('Frequency')
plt.title('Top 10 Brands and Others')
plt.xticks(rotation=45)  # Rotate labels for better readability

# Show the plot
plt.show()
```
